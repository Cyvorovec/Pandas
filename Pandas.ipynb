4.5.2) Создать DataFrame в котором присутствует столбец, отражающий разницу в объемах торгов по WAT и ALTR в одинаковые дни и содержит исходные данны об объеме трогов этими акциями.

```
# переиндексация dataframe
WAT = zd2 [zd2 ['Index'] == 'WAT']
WAT.reset_index(inplace=True)
ALTR.reset_index(inplace=True)
```

```
k=len(WAT.Index)
n=len(ALTR.Index)

m = pd.DataFrame ({
    'Amount1':[],
    'Amount2':[],
    'Diff':[],
})

for i in range (k):
  for j in range (n):
    if WAT.loc[i , 'Date'] == ALTR.loc [j , 'Date']:
      m.loc[i, 'Diff']=ALTR.loc [j , 'Amount'] - WAT.loc[i , 'Amount']
      m.loc[i, 'Amount1']=ALTR.loc[j , 'Amount']
      m.loc[i, 'Amount2']=WAT.loc [i , 'Amount']
      break

print (m)
```

Добавить столбец с расшифровкой названия тикера используя данные из файла sp_data2.csv . В случае нехватки данных об именах тикеров корректно обработать их.

```
pi = pd.read_csv("sp_data2.csv",
                  sep = ';',
                  names = ['Tiker','Desc','Percentage'])
#122574
k=len(zd2.Index)
n=len(pi.Tiker)

zd2['PIZ']='No data'

for i in range (k):
  for j in range (n):
    if zd2.loc[i , 'Index'] == pi.loc [j , 'Tiker']:
      zd2.loc[i, 'PIZ']=pi.loc [j , 'Desc']
      break
print (zd2)
```
